```python
import json
import datetime

schedule_list=[]
output=[]
with open('output_data.json',encoding='utf-8',errors='ignore') as f:
    data = json.load(f)

cnt = 0
currenttime = data['CURRENTTIME']
lot_cnt ={}
for lot in data['data_dict']['schedule']:
    item = data.get('data_dict').get('schedule').get(lot)
    for a in item['bucket_list']:
        key = a['data_key']
        lot_item=data['data_dict']['bucket_ref'].get(key)
        s = {}
        if lot_item != None:
            data_key =a['name']+ '-' + lot_item['op']['id']+'-'+str(lot_item['op']['data_dict']['layer_code'])
            if data_key in lot_cnt:
                lot_cnt[data_key] = int(lot_cnt[data_key]+ 1) 
            else:
                lot_cnt[data_key] =1
            lot = data_key+"-"+str(lot_cnt[data_key])
#             print(lot)
            s['LOT'] = lot
            s['ORDER_ID'] = a['name']
            s['OPE_ID'] = lot_item['op']['id']
            s['LAYER'] = lot_item['op']['data_dict']['layer_code']
            s['PRODUCT_ID'] = lot_item['order']['pd_id']
            s['QTY'] = lot_item['op']['data_dict']['PANEL_QTY']
            s['MACHINE_ID'] = item["id"]
            s['START_TIME'] = datetime.datetime.fromtimestamp(a["start"]+currenttime).strftime('%Y/%m/%d %H:%M:%S')
            s['END_TIME'] = datetime.datetime.fromtimestamp(a["finish"]+currenttime).strftime('%Y/%m/%d %H:%M:%S')
            s['CR'] = lot_item['order']['CR']
            s['SHIPPING_DATE'] = datetime.datetime.fromtimestamp(lot_item['order']['DD']+currenttime).strftime('%Y/%m/%d %H:%M:%S')
            s['CUTTING_TIME'] = datetime.datetime.fromtimestamp(a["start"]+currenttime+1200).strftime('%Y/%m/%d %H:%M:%S')
            s['DAYPLAN_NO'] = str((datetime.date.today().strftime('%Y%m%d')))
            
            schedule_list.append(s)

with open('output.json', 'w') as f:
    json.dump(schedule_list, f)
```

```python
# 機台
import json

output=[]
with open('output_data.json',encoding='utf-8',errors='ignore') as f:
    data = json.load(f)

for group in data['data_dict']['mcg_dict']:
    s = {}
    item = data.get('data_dict').get('mcg_dict').get(group)[0]['mc_list']
    
    for machine in item:
#         print(machine)
        s['GROUP_ID'] = machine['data_dict']['id']
        s['GROUP_NAME'] = machine['data_dict']['mcg_name']
        s['MACHINE_ID'] = machine['id']
        s['MACHINE_NAME'] = machine['id']
    output.append(s)
# print(output)
with open('mc_list.json', 'w', encoding="utf-8") as f:
    json.dump(output, f, indent=4, ensure_ascii=False)
```
